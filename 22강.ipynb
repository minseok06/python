# 준원이 바보

```python
sns.set_theme()
plt.figure(dpi=200)
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill', height=0.08)
```

# 결재금액의 20%의 정도의 팁을 낸다.
1. total_bill의 따라 tip이 높아지는 것을 봐서 정비례 한 것 같지만, 많이 먹고 조금내는 양아치도 있다.
2. 보면 조금 먹고 많이내는 천사도 있다.
3. x축과 y축은 눈금선 같은 비슷한 역할을 해준다.

```python
sns.rugplot(data=tips, x='total_bill', y='tip')
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

# 종별 펭귄 몸무게의 따른 성별 분포율
1. 첫번째를 보면 성별 분포는 비슷하고 몸무게가 다양 한 걸 봐서 평범한 펭귄이다.
2. 두번째는 개체 수가 낮을 걸 봐서 희귀한 펭귄이다.
3. 세번째는 몸무개가 많이 나가는 걸 봐서 자이언트 펭귄이다.

```python
sns.displot(data=penguins, x='body_mass_g', col='species', hue='sex')
```

```python
sns.displot(data=tips, x='tip', col='time')
sns.rugplot(data=tips, x='tip')
```
