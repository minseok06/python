# 기존의 변수를 건드리지 않고 새로운 변수로 만든다

```python
df = pd.read_csv('data/employees.csv', parse_dates = ['Start Date','Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')

mask1 = df['Gender']=='Male'
mask2 = df['Team']=='Marketing'

df[mask1 & mask2].head()
```

```python
female = df['Gender']=='Female'
finance = df['Team']=='Finance'
```

```python
df[female | finance]
```

```python
male = df['Gender']=='Male'
Cli = df['Team']=='Client Services'
year = df['Start Date']>'2016-06-01'
```

```python
df[(male & Cli) | year].head()
# () 소가로까지 해줘야 안전 빵으로 갈 수 있다.
```

```python
L = df['Team']=='Legal'
P = df['Team']=='Product'
S = df['Team']=='Sales'

df[L | P | S]
```
