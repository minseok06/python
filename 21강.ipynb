```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# load_dataset()

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Scatterplots

```python
sns.set_theme()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
tips.plot.scatter('total_bill','tip', color='b')
```

```python
tips.plot(kind='scatter', x='total_bill', y='tip', color='b')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
# hue를 sex로 구분하여 나눠줬다.
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
# hue를 smoker로 구분하여 나눠줬다.
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
# hue를 day로 구분하여 나눠줬다.
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='tip', style='sex')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.plot(x='year',y='passengers')
```

```python
sns.lineplot(data = flights, x = 'year' , y='passengers')
```

```python
import matplotlib.pyplot as plt
flights.groupby('year')['passengers'].mean().plot()
```

```python
sns.lineplot(data = flights, x='year', y='passengers', hue='month')
```

```python
trips = sns.load_dataset('taxis', parse_dates=['pickup','dropoff'])
trips.head()
```

```python
# hour 컬럼 만들기
```

```python
trips['hour'] = trips['pickup'].dt.hour
trips['hour']
```

```python
trips
```

```python
trips.info()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', kind='scatter', 
            col='day', row='sex', height=1.5, aspect=1.5, estimator=None)
```

```python
# Relplot : 데이터를 다양한 기준에 따라 따로 그래프화 하는 메소드
# col, row의 차이를 잘 알아야한다.
```

```python
sns.histplot(data=tips, x='tip')
```

```python
sns.histplot(data=tips, y='tip')
```

```python
sns.relplot(data=trips, x='hour', y='total', hue='payment', kind='line',
            col='pickup_borough')
```
