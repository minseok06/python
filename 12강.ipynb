```python
dj = [16,6,19,30]
so = [29,15,15,23]
db = ["강원","경기","인천","충청"]
# 변수 데이터들

plt.figure(figsize=(8,6))
# 그래프 사이즈 비율
plt.style.use('fivethirtyeight')
# 그래프 배경 스타일
plt.bar(db, dj, bottom=so, color="#FFA100", label="어제 적설량")
plt.bar(db, so, color="#00A6ED",label="내일 예상 적설량")
plt.legend()
# 라벨 표시 박스
plt.ylabel("Inches")
plt.title("적설량 보고서")
# 제목
```

```python
labels = [
    "잠자기","잘 준비","저녁 먹기","방과후","학교(5~6교시)","점심먹기","학교(1~4교시)","학교 가기"
]
time = [
    90, 10, 10, 30, 20, 20, 30, 10
]
colors = ["#355070", "#00A6ED","#eaacBb","#FFA100","#e55b6f","#eaacBb", "#e55b6f", "#eaacBb"]
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    time,
    labels=labels,
    colors=colors,
    explode=(0, 0, 0.1,0,0,0.1,0,0.1),
    autopct='%.0f%%'
)
plt.title("하루 시간 편성표")
plt.show()
```
