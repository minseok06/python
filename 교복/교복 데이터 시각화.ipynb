```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 라이브러리 불러오기 및 별명 지정
plt.rc('font', family = 'Malgun Gothic')
# 한글폰트 적용하기

file = '0급-기밀파일.csv' # 파일 불러오기
df = pd.read_csv(file, encoding='cp949') # 데이터 프레임으로 변환
df.head(1) # 출력
```

```python
df_ok = df[df['교복 착용']=='네']['교복 필요성'].value_counts()
df_no = df[df['교복 착용']=='아니오']['교복 필요성'].value_counts()
```

```python
df['교복 필요성'].value_counts()
```

```python
df_ok = df[df['교복 착용']=='네']['교복 필요성'].value_counts()
df_no = df[df['교복 착용']=='아니오']['교복 필요성'].value_counts()
l = ['필요해요', '불필요해요']

plt.plot(l, df_ok, color='royalblue', label='예')
plt.plot(l, df_no, color='hotpink', label='아니오')
plt.xlabel('교복 필요성')
plt.legend(title='교복 착용 여부')
```

```python
k1 = df[df['학년'] == '1학년']['교복 착용'].value_counts()
k2 = df[df['학년'] == '2학년']['교복 착용'].value_counts()
k3 = df[df['학년'] == '3학년']['교복 착용'].value_counts()
```

```python
print(k1)
print(k2)
print(k3)
```

```python
explode=[0.01,0.01]
# 파이 여백
labels = ['예', '아니오']
colors = ['royalblue','hotpink']
# 라벨 이름

plt.pie(k1, labels=labels,
        autopct='%1.1f%%', # 소수 첫째자리 표시
        counterclock=False, # 파이 방향
        startangle=90, # 시작 각도
        explode=explode,
        colors=colors
       )

plt.legend(loc = (0.8,0.8), title = '1학년')
plt.show()
```

```python
sns.set_palette('Set3')
# seaborn 스타일

explode=[0.01,0.01]
# 파이 여백
labels = ['예', '아니오']
# 라벨 이름

plt.pie(df['교복 착용'].value_counts(), labels=labels,
        autopct='%1.1f%%', # 소수 첫째자리 표시
        counterclock=False, # 파이 방향
        startangle=90, # 시작 각도
        explode=explode
       )

plt.legend(loc = (0.8,0.8), title = '교복 잘 입나요?')
plt.show()
```

```python
sns.set_palette('pastel')

l = ['네', '아니오']
k = df['교복 착용'].value_counts()
plt.bar(l, k, label='교복착용')
plt.legend()
plt.show()
```
