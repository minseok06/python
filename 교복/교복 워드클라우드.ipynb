```python
from wordcloud import WordCloud
from string import digits
import re
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

font_path = 'C:/Windows/Fonts/malgun.ttf' # 글꼴 지정
plt.rc('font',family='Malgun Gothic') # 그래프 한글 패치
```

```python
text = '교복 필요성.txt'

with open(text, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines:
        text += line

text = re.sub(r'[0-9]+','',text)
text = re.sub(r'수','',text)
text = re.sub(r'때문에','',text)
text = re.sub(r'입을','',text)
text = re.sub(r'txt','',text)
text = re.sub(r'거','',text)
text = re.sub(r'잘','',text)
text = re.sub(r'각','',text)
text = re.sub(r'할','',text)
text = re.sub(r'을','',text)
text = re.sub(r'좀','',text)
text = re.sub(r'있','',text)
text = re.sub(r'때만','',text)
text = re.sub(r'으니까','',text)
text = re.sub(r'생합니다','',text)
text = re.sub(r'교복이','',text)
text = re.sub(r'좋지','',text)
text = re.sub(r'같은','',text)

wc = WordCloud(font_path=font_path, background_color='white',
               prefer_horizontal=True, width=600, height=400)

wc.generate(text)
plt.figure(figsize=(8,8))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
