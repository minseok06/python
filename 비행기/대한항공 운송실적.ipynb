# 항공사 대한항공 추출하기

```python
korean_air = airline_2018[airline_2018['항공사명'] == '대한항공']
korean_air.head()
```

```python
korean_air.groupby(['년월','항공사명'])[['유임승객(명)']].sum().plot()
plt.title('2018년 월별 대한항공 탑승객 통계')
plt.xticks(rotation=20)
plt.plot()
```
