```python
import numpy as np
import pandas as pd
```

```python
data1 = [['A0','B0','C0','D0'],['A1','B1','C1','D1'],['A2','B2','C2','D2'],['A3','B3','C3','D3']]
df1 = pd.DataFrame(data1,index='0 1 2 3'.split(),columns='A B C D'.split())
data2 = [['A4','B4','C4','D4'],['A5','B5','C5','D5'],['A6','B6','C6','D6'],['A7','B7','C7','D7']]
df2 = pd.DataFrame(data2,index='4 5 6 7'.split(),columns='A B C D'.split())
data3 = [['A8','B8','C8','D8'],['A9','B9','C9','D9'],['A10','B10','C10','D10'],['A11','B11','C11','D11']]
df3 = pd.DataFrame(data3,index='8 9 10 11'.split(),columns='A B C D'.split())
```

### Concatenation (concat) 분산된 데이터를 합치는 매소드

```python
pd.concat([df1,df2,df3]).transpose()
```

```python
pd.concat([df1,df2,df3],axis=1)
```

### axis = 0 행방향으로 동작, axis =1 열방향으로 동작

```python
left = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                   'A': ['A0','A1','A2','A3'],
                   'B': ['B0','B1','B2','B3']})

right = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                   'C': ['C0','C1','C2','C3'],
                   'D': ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

### Left와 right 두가지 데이터 중 내용이 같은 key를 기준으로 두 데이터를 합쳐서 보여줌. 
### Inner라는 조건은, 두개의 데이터 셋에 공통된 부분만 합하여 준다는 뜻. 

```python
right = pd.DataFrame({'key': ['K0','K1','K2','K4'],
                   'C': ['C0','C1','C2','C3'],
                   'D': ['D0','D1','D2','D3']})
```

```python
right
```

```python
pd.merge(left, right, how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

### 합치려는(Merge 하려는) 데이터셋의 key를 모두 가져와서, 기준 컬럼으로 사용하고, 해당 key에 데이터가 없는 부분은 NaN으로 출력되도록 한다. 

```python
left1 = pd.DataFrame({'key1': ['K0','K0','K1','K2'],
                      'key2': ['K0','K1','K0','K1'],
                           'A': ['A0','A1','A2','A3'],
                           'B': ['B0','B1','B2','B3']})

right1 = pd.DataFrame({'key1': ['K0','K1','K1','K2'],
                               'key2': ['K0','K0','K0','K0'],
                               'C': ['C0','C1','C2','C3'],
                               'D': ['D0','D1','D2','D3']})
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, on=['key1','key2'])
```

### INNER라는 조건은 DEFAULT이기 때문에 쓰지 않아도 된다

```python
pd.merge(left1, right1, how='outer', on=['key1', 'key2'])
```
