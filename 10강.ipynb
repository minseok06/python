```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

Plt에 matplotlib 을 저장했고, matplotlib에 있는 plot이라는 메소드를 통해 어떤 상관관계가 있는 그래프를 그릴 것인지를 지정하고, 해당 그래프의 색깔을 지정한다.

```python
plt.plot(y, x, 'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title Here')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(1,3,1)
plt.plot(x, y, 'r--')
plt.subplot(1,3,1)
plt.plot(y, x, 'g*-')
```

```python
plt.subplot(1,3,2)
plt.plot(x,y, 'r--')
plt.subplot(1,3,1)
plt.plot(y,x, 'g*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.3, 0.9])

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Figure라는 메소드는 그래프의 크기와 가로 세로 비율을 정할 수 있도록 해준다.
# axes는 축이다.
# fig.add_axes([왼쪽축에서의 거리, 바닥에서의 거리, 그래프의 폭, 그래프의 높이])
# 위 add_axes 안에 들어갈 수 있는 숫자의 범위는 0부터 1까지 가능
# 
```

```python
fig2 = plt.figure()

axes = fig2.add_axes([1, 0, 0.4, 0.3])

axes.plot(y, x, 'r')
axes.set_xlabel('X_label_axes2')
axes.set_ylabel('Y_label_axes2')
axes.set_title('Axes 2 Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.5, 0.3])

axes.plot(x, y, 'b')
axes.set_xlabel('X_label_axes2')
axes.set_ylabel('Y_label_axes2')
axes.set_title('Axes 2 Title')
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.7])
axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# 이것이 있어야 한다.
```

```python
# 반복문을 사용해서 해보자
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
fig
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.3, 0.6])
axes2 = fig.add_axes([0.5, 0.1, 0.3, 0.6])

axes.plot(x, y, 'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
axes2.plot(x, y, 'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('title')
```
