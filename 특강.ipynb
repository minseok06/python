```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('data/subway202306.csv')
data = csv.reader(f)
next(data)
next(data)
result = []

for row in data :
    row[4:] = map(int, row[4:])
    result.append(row[7])

result.sort(reverse=True)
    
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.bar(range(len(result)), result, label = '5시 하차')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('data/subway202306.csv')
data = csv.reader(f)
next(data)
next(data)
result = []

for row in data :
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))

result.sort(reverse=True)
    
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.bar(range(len(result)), result, label = '7~9시 승차인원')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('data/subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

for row in data :
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > mx:
        mx = sum(row[10:15:2])
        mx_station = row[3]+'('+row[1]+')'

print(mx_station, mx)
```
