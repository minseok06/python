```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = pd.read_csv("data/Salaries.csv", low_memory=False)
s.head()
```

```python
s["BasePay"]=pd.to_numeric(s["BasePay"], errors="coerce").fillna(0)
s["OvertimePay"]=pd.to_numeric(s["OvertimePay"], errors="coerce").fillna(0)
s["OtherPay"]=pd.to_numeric(s["OtherPay"], errors="coerce").fillna(0)
s["EmployeeName"]=pd.to_numeric(s["EmployeeName"], errors="coerce").fillna(0)
df = s[["EmployeeName", "BasePay", "OvertimePay", "OtherPay"]]
df.set_index("EmployeeName", inplace=True)

df.head(10).plot(kind="bar", stacked=True)
```

### 만약 숫자로 변경할 수 없는 데이터라면 기존 데이터를 지우고 
### NaN으로 설정하여 반환하는 메소드 : errors="coerce"
#### fillna(0) : NaN으로 기록된 부분을 숫자 0으로 반환하라는 메소드 

```python
s[["BasePay","OvertimePay","OtherPay"]].head(10).plot(kind="bar")
plt.xlabel("EmployeeName")
```

Sorting => 정렬하다.
sort_values => 값의 순서대로 정렬.

```python
b = s["BasePay"].sort_values(ascending=False).head()
b.plot(kind="bar")
```
